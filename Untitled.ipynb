```python
#Gerekli kütüphanelerin tanımlanması
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#"csv" formatındaki veri setinin "veri_seti" değişkenine atanması
veri_seti = pd.read_csv("tiktok_dataset.csv")

#"veri_seti" değişkenindeki verilerin projede kullanılabilmesi için "tiktok" değişkenine atanması.
tiktok = pd.DataFrame(veri_seti)
```

```python
#Veri seti içerisinde yer alan sütunların boş değer sayılarının görüntülenmesi
tiktok.isnull().sum()
```

```python
#"claim_status" sütünundaki kategorilerin tekrar etme durumunun görüntülenmesi
print( tiktok["claim_status"].value_counts() )
```

```python
#Boş değerleri doldurma işlemi
#"opinion" sayısı az olduğu için boş satırları bu şekilde doldurdum.
tiktok["claim_status"] = tiktok["claim_status"].fillna("opinion") #opinion sayısı az olduğu için boş satırları bu şekilde doldurdum.
```

```python
#"video_transcription_text" sütunundaki ilgili kategorilerin tekrar etme durumunun görüntülenmesi
print( tiktok["video_transcription_text"].value_counts() )
```

```python
#"video_transcription_text" sütununa ait boş değerlerin doldurulması
#Boş değerleri "no-transcription" ile doldurdum.
tiktok["video_transcription_text"] = tiktok["video_transcription_text"].fillna( "no-transcription" )
```

```python
#"video_view_count" sütunundaki kategorilere ait verilerin kaç kez tekrarlandığının görüntülenmesi
print (tiktok["video_view_count"].value_counts() )
```

```python
#"video_view_count" sütunundaki boş olmayan değerlerin ortalaması
print( "Görüntelenme Ortalama: ", float( tiktok["video_view_count"].mean() ) )
```

```python
#"video_view_count" sütunundaki boş değerlerin doldurulması (ortalama değer ile dolduruldu)
tiktok["video_view_count"] = tiktok["video_view_count"].fillna(tiktok["video_view_count"].mean())
```

```python
#"video_like_count" sütunundaki kategorilere ait boş değerlerin listelenmesi
print (tiktok["video_like_count"].value_counts() )
```

```python
#"video_like_count" sütunundaki boş olmayan değerlerin modunun bulunması
print( "Beğenme sayısının modu: ", tiktok["video_like_count"].mode()[0] )
```

```python
#"video_like_count" sütunundaki boş değerlerin bulunan mod değeriyle doldurulması
tiktok["video_like_count"] = tiktok["video_like_count"].fillna(tiktok["video_like_count"].mode()[0])
```

```python
#"video_share_count" sütunundaki kategorilere ait boş değerlerin sayılarının görüntülenmesi
print (tiktok["video_share_count"].value_counts() )
```

```python
#"video_share_count" sütununa ait boş olmayan ver,
print( "Paylaşım sayısının modu: ", tiktok["video_share_count"].mode()[0] )
```

```python
#"video_share_count" sütunundaki boş verilerin bulunan mod değeriyle doldurulması
tiktok["video_share_count"] = tiktok["video_share_count"].fillna( tiktok["video_share_count"].mode()[0] )
```

```python
#"video_download_count" sütunundaki boş verilerin kategorilere göre sayısı
print (tiktok["video_download_count"].value_counts() )
```

```python
#"video_download_count" sütunundaki boş olmayan verilerin ortalamasının bulunması
print( "İndirme sayısının ortalaması: ", float( tiktok["video_download_count"].mean() ) )
```

```python
#"video_download_count" sütunundaki boş verilerin bulunan ortalama ile doldurulması
tiktok["video_download_count"] = tiktok["video_download_count"].fillna(tiktok["video_download_count"].mean())
```

```python
#"video_comment_count" sütunundaki boş değerlerin kategorilere sayısının görüntülenmesi
print( tiktok["video_comment_count"].value_counts())
```

```python
#"video_comment_count" sütunundaki boş olmayan değerlerin ortalamasının bulunması
print("Yorum sayısının ortalaması: ",float(tiktok["video_comment_count"].mean() ) )
```

```python
#"video_comment_count" sütunundaki boş değerlerin bulunan ortalama değer ile doldurulması
tiktok["video_comment_count"]=tiktok["video_comment_count"].fillna(tiktok["video_comment_count"].mean() )
```

```python
#Histogram Grafiği
#"video_duration" sütunun frekans dağılımı
plt.figure(figsize=(10, 6)) # Grafik boyutu

plt.hist(tiktok["video_duration_sec"], color="cornflowerblue") # Histogram grafiğinin çizilmesi

plt.title("Frekans Dağılım Grafiği") #GrafikBaşlığı
plt.xlabel("Süre") # Satırın başlığı
plt.ylabel("Frekans") # Sütunun başlığı

plt.show() #Grafiğin gösterilmesi
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))  #Grafik boyutu

ban_counts = tiktok["author_ban_status"].value_counts()  #"author_ban_status"  sütununa ait veriler
ban_labels = ban_counts.index  #Sütun başlıkları

# Pasta grafiğinin çizilmesi
plt.pie(
    ban_counts, 
    labels=ban_labels, 
    autopct="%1.1f%%",  #Yüzde değerlerinin gösterimi
    colors=["mediumseagreen", "orchid", "indianred"], #("3 kategori olduğunu bildiğimden renkleri bu şekilde tanımladım")
    startangle=140  #Başlangıç açısı
)

plt.title("Yasaklama Dağılım Grafiği")  #Grafik başlığı
plt.show()  #Grafiğin gösterimi
```

```python
plt.figure( figsize=(10,6) ) #Grafik boyutu

verified_counts = tiktok["verified_status"].value_counts() #"verified_status" sütununun sayıları
verified_labels = tiktok["verified_status"].value_counts().index # verified_status" sütununun başlıklarının gösterimi

plt.xticks(rotation=90) #Başlıkların 90 derece çevrilmesi

plt.title("Doğrulama Durum Grafiği")  #Grafik başlığı
plt.bar(verified_labels, verified_counts, width=0.5, color="crimson") #Bar grafiğinin çizilmesi
plt.show() #Grafiğin gösterilmesi
```

```python
plt.figure(figsize=(12, 6))  #Grafik boyutu

#Kutu grafiğinin çizilmesi
sns.boxplot(data=tiktok, x="author_ban_status", y="video_view_count", palette="rocket") #Kontrol edilecek değişkenlerin verilerin alınması

plt.xlabel("Ban Durumu")  #Satır ismi
plt.ylabel("Görüntülenme")  #Sütun ismi
plt.title("Ban Durumuna Göre Görüntülenme Sayısı") #Grafik başlığı
plt.show() #Grafiğin görüntülenmesi
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(tiktok["video_like_count"], tiktok["video_download_count"], color="coral", alpha=0.6) #Seçilen sütunlar ve grafiğin çizilmesi
plt.title("Beğeni ve İndirme Sayısı Arasındaki İlişki") #Grafik başlığı
plt.xlabel("Beğeni Sayısı") #Satır ismi
plt.ylabel("İndirme Sayısı") #Sütun ismi
plt.show() #Grafiğin gösterilmesi
```

```python
# Sayısal sütunların seçimi
numeric_columns = tiktok[["video_view_count", "video_like_count", "video_comment_count"]]

# Korelasyon matrisinin hesaplanması
correlation_matrix = numeric_columns.corr()

# Korelasyon matrisinin yazdırılması
print(correlation_matrix)

# Korelasyon grafiğinin görselleştirilmesi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="YlOrBr", fmt='.2f', vmin=-1, vmax=1)
plt.title("Korelasyon Matrisi")
plt.show()
```
